**Vectorization**
In deep learning, you deal with massive and complex big data sets. In this case, a non-computationally efficient function can become a hurdle, and our model may take ages to train. The key concept for efficeint computing is to avoid as many for loops as possible. Python has a powerful function, known as vectorization. Let's illustrate this with an example.

```
import numpy as np
import time
a = np.arange(1000)
b = np.arange(1000)
# pure Python outer product implementation
tic = time.process_time()
outer_product = np.zeros((1000, 1000))
for i in range(len(a)):
    for j in range(len(b)):
        outer_product[i][j]= a[i] * b[j]
toc = time.process_time()
print("python_outer_product = "+ str(outer_product))
print("Time = "+str(1000*(toc - tic ))+"ms\n")
# Numpy outer product implementation
n_tic = time.process_time()
outer_product = np.outer(a, b)
n_toc = time.process_time()
print("numpy_outer_product = "+str(outer_product));
print("Time = "+str(1000*(n_toc - n_tic ))+"ms")
```

```
x1 = [9, 2, 5, 0, 0, 7, 5, 0, 0, 0, 9, 2, 5, 0, 0]
x2 = [9, 2, 2, 9, 0, 9, 2, 5, 0, 0, 9, 2, 5, 0, 0]

### VECTORIZED DOT PRODUCT OF VECTORS ###
tic = time.process_time()
dot = np.dot(x1,x2)
toc = time.process_time()
print ("dot = " + str(dot) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

### VECTORIZED OUTER PRODUCT ###
tic = time.process_time()
outer = np.outer(x1,x2)
toc = time.process_time()
print ("outer = " + str(outer) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

### VECTORIZED ELEMENTWISE MULTIPLICATION ###
tic = time.process_time()
mul = np.multiply(x1,x2)
toc = time.process_time()
print ("elementwise multiplication = " + str(mul) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

```

**Broadcasting**
 It is also a great feature when working with numpy arrays and neural netrworks. It allows  array arithmetic between arrays with a different shape or size. Let's see this example.

```
# scalar and one-dimensional
import numpy as np
a = np.array([1, 2, 3])
print(a)
b = 2
print(b)
c = a + b
print(c)
```

```
# # scalar and two-dimensional
from numpy import array
A = array([[4, 5, 7], [1, 2, 3]])
print(A)
b = 2
print(b)
C = A + b
print(C)
```

**SIGMOID FUNCTION**

```
import matplotlib.pyplot as plt
import numpy as np
import math
import random 
```

```
x=np.linspace(-10,10,10)
x
```

```
def sigmoid(x):
  sigmoid=[]
  for i in x:
    sigmoid.append(1/(1+np.exp((-1)*i)))

  return sigmoid
```

```
ans = sigmoid(x)
```

```
plt.plot(x,ans);
plt.xlabel("x")
plt.ylabel("sigmoid")
```

**SINE FUNCTION**

```
def sin(x):
  sine=0
  for i in range(50):
    sign=math.pow(-1,i)
    pi=math.pi
    a=x*(pi/180)
    sine = sine+(sign*(a**(2.0*i+1))/math.factorial(2*i+1))
  return sine
```

```
x=float(input("enter degrees : "))
print(sin(x))
```

**IMAGE TO VEC**

```
import numpy as np
def image_to_vec(image):
  length = image.shape[0]
  height = image.shape[1]
  depth = image.shape[2]
  image = image.reshape(length*height*depth,1)
  return image
```

```
image = np.random.rand(2,2,4)
print(image_to_vec(image))
```

**TAN H**

```
def tanh(x):
  ans = (np.exp(x) - np.exp(-x)) / np.exp(x) + np.exp(-x)
  return ans
```

```
x = np.linspace(-10,10,10)
ans = tanh(x)
ans
```

```
import matplotlib.pyplot as plt
plt.plot(x,ans,color='blue',marker='o')
```

**SOFTMAX**

```
def softmax(x):
  return np.exp(x-np.max(x))/sum(np.exp(x))
```

**LEAKY RELU**

```
def leaky_relu(x):
  if(x>0):
    return x
  else:
    return 0.01*x
```

```
x=[-10,-6,-4,-3,-2,0,1,2,3,4,5,6]
```

```
ans=[]
for i in x:
  ans.append(leaky_relu(i))
```
